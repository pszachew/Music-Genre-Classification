```python
import numpy as np
import os
import librosa
import pandas as pd
```

```python
X = []
labels = []
initial_path = '/content/gdrive/MyDrive/Data/TinySOL1s'
for category in os.listdir(initial_path):
    for instrument in os.listdir(initial_path+"/"+category):
        for file in os.listdir(initial_path+"/"+category+"/"+instrument+"/ordinario"):
          src, sr = librosa.load(initial_path+"/"+category+"/"+instrument+"/ordinario"+"/" + file)
          mfcc = librosa.feature.mfcc(y=src, sr = sr, n_mfcc=20)
          dmfcc = mfcc[:, 1:] - mfcc[:, :-1]
          ddmfcc = dmfcc[:, 1:] - dmfcc[:, :-1]
          X.append(np.concatenate((np.mean(mfcc, axis=1), np.std(mfcc, axis=1),
                           np.mean(dmfcc, axis=1), np.std(dmfcc, axis=1),
                           np.mean(ddmfcc, axis=1), np.std(ddmfcc, axis=1))
                          , axis=0))
          labels.append(instrument)
```

```python
X_df = pd.DataFrame(X)
X_df
```

```python
labels_df = pd.DataFrame(labels)
labels_df
```

```python
labels_df.to_csv("/content/gdrive/MyDrive/csv_data/TinySOL1s/labels_mfcc.csv", index=False, header=False)
```
