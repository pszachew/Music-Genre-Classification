```python
from pydub import AudioSegment
```

```python
path_in = '/content/gdrive/MyDrive/MagnaTagATune/mp3'
path_out = '/content/gdrive/MyDrive/MagnaTagATune/wav'
path_out_emb = '/content/gdrive/MyDrive/Embedded Music/MagnaTagATune'
dirlist = ['2', '3', '4', '5']
for dir in dirlist:
  for file in os.listdir(path_in+'/'+dir):
      file_out = ''.join(file.split('.')[:-1])+".wav"
      AudioSegment.from_mp3(path_in+'/'+dir+'/'+file).export(path_out+'/'+dir+'/'+file_out ,format="wav")
      openl3.process_audio_file(path_out+'/'+dir+'/'+file_out, output_dir=path_out_emb+'/'+dir, hop_size=1, embedding_size=512)
```
